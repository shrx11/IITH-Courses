```python
import pandas as pd
import numpy as np
import sklearn
import sklearn.ensemble
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_hastie_10_2
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
def prepro(df):
  df['Scenes'] = df['Scenes'].str.cat(df['Sl. No.'], sep ="/")
  df.drop(['Sl. No.'],axis=1,inplace=True)
  df['Scenes'] = df['Scenes'].map(lambda x: str(x)[6:])
  return df
```

```python
df22=prepro(df22)
df25=prepro(df25)
df26=prepro(df26)
df27=prepro(df27)
df30=prepro(df30)
df31=prepro(df31)
df32=prepro(df32)
df33=prepro(df33)
df36=prepro(df36)
df37=prepro(df37)
```

```python
ep22=df22.merge(EP22_vf,left_on='Scenes', right_on='scene')
ep25=df25.merge(EP25_vf,left_on='Scenes', right_on='scene')
ep26=df26.merge(EP26_vf,left_on='Scenes', right_on='scene')
ep27=df27.merge(EP27_vf,left_on='Scenes', right_on='scene')
ep30=df30.merge(EP30_vf,left_on='Scenes', right_on='scene')
ep31=df31.merge(EP31_vf,left_on='Scenes', right_on='scene')
ep32=df32.merge(EP32_vf,left_on='Scenes', right_on='scene')
ep33=df33.merge(EP33_vf,left_on='Scenes', right_on='scene')
ep36=df36.merge(EP36_vf,left_on='Scenes', right_on='scene')
ep37=df37.merge(EP37_vf,left_on='Scenes', right_on='scene')
```

```python
videodf=pd.concat([ep22,ep25,ep26,ep27,ep30,ep31,ep32,ep33,ep36,ep37])
videodf.Label.replace(['humor', 'non-humor'], [1, 0], inplace=True)
videodf.drop(['Scenes','scene','Start_time','End_time','Utterance'],inplace=True,axis=1)

videodf['Speaker'] = videodf['Speaker'].str.split(',').str[0]  

videodf['Speaker']= videodf['Speaker'].astype('category')
videodf['Speaker'] = videodf['Speaker'].cat.codes
```

```python
videodf.dropna(axis=0,inplace=True)
#videodf.to_csv("/content/drive/MyDrive/IR project/videodf.csv",sep=',')
videodf
```

```python
train_df, test_df = train_test_split(videodf, random_state=91,test_size=0.2)
```

```python
#only emotion probabilities
train_df = train_df.drop(['fear','disgust','happiness','surprise','neutral','disgust.1','happiness.1','surprise.1','neutral.1'],axis=1)
test_df = test_df.drop(['fear','disgust','happiness','surprise','neutral','disgust.1','happiness.1','surprise.1','neutral.1'],axis=1)
```

```python
train_X_df=train_df.drop(['Label'],axis=1)
test_X_df=test_df.drop(['Label'],axis=1)

train_X=train_X_df.to_numpy()
train_Y=train_df['Label'].to_numpy()
test_X=test_X_df.to_numpy()
test_Y=test_df['Label'].to_numpy()
```

```python
clf = GradientBoostingClassifier(n_estimators=800, learning_rate=0.2,criterion='friedman_mse',max_depth=5, random_state=10,max_features=50).fit(train_X, train_Y)
print(clf.score(test_X, test_Y))
print(metrics.precision_score(test_Y, clf.predict(test_X)))
print(metrics.recall_score(test_Y, clf.predict(test_X)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_Y, clf.predict(test_X)))
print(confusion_matrix(test_Y, clf.predict(test_X)))
```
